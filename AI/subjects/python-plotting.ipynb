# Kreslení v Pythonu

budeme používat knihovnu Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Rychlé kreslení

```python
x = np.linspace(0, 2*np.pi)
fx = np.sin(x)
plt.plot(x, fx)
```

# Jednoduchý příklad
---
který není úplně jednoduchý
nejprve pojmy figure, axes, axis
jde to i jednodušeji, ale takhel se nám budou lépe kreslit složitější výstupy

```python
fig = plt.figure()   # prazdny obrazek
fig, ax = plt.subplots()  # jedny axes v obrazku, ale lze zadat i mrizku plt.subplots(2, 2)
ax.plot(x, fx)  # vykresleni bodu
```

## Stylování
---

### Čáry


```python
x = np.linspace(0, 2*np.pi)
fx = np.sin(x)
gx = np.cos(x)
fig = plt.figure()   
fig, ax = plt.subplots()  
ax.plot(x, fx, lw=3, ls="--") # ls = linestyle, lw = linewidth
ax.plot(x, gx, lw=5, ls=":")
```

### Barvičky
[dokumentace barvy](https://matplotlib.org/stable/tutorials/colors/colors.html)

```python
fig = plt.figure()   
fig, ax = plt.subplots()  
ax.plot(x, fx, color="coral") 
ax.plot(x, gx, color="olive")
```

### Markery

```python
x = np.arange(12)
y = np.exp(x)
fig = plt.figure()   
fig, ax = plt.subplots()
ax.plot(x, y, marker="o", ls="-")   # stejne jako ax.plot(x, y, "o-") :) 


```

## Scatter
----

```python
from numpy import random  

x = random.rand(40)
y = random.rand(40)

fig = plt.figure()   
fig, ax = plt.subplots()  
ax.scatter(x, y, s=50, fc="blue", ec="k")   # s je velikost, fc=facecolor, ec=edgecolor
```

```python
x = random.rand(40)
y = random.rand(40)

fig = plt.figure()   
fig, ax = plt.subplots() 

ax.set_aspect("equal")   # <-------- čtvercové souřadnice


ax.scatter(x, y, s=100, fc="blue", ec="k")   # s je velikost, fc=facecolor, ec=edgecolor
```

## Histogram
---

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots()

ax.hist(x, 20) # 20 bins
```

## Koláčový graf
---

```python
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1i%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

## Heatmapa

## Popisky
---

```python
x = np.linspace(0, 2*np.pi)
fx = np.sin(x)
gx = np.cos(x)

fig = plt.figure()   
fig, ax = plt.subplots()  
ax.plot(x, fx, color="coral", label="SIN") 
ax.plot(x, gx, color="olive", label="COS")
ax.legend()  # vyvolání legenda
```

### Popisky os a obrázku

```python
fig = plt.figure()   
fig, ax = plt.subplots()  
ax.plot(x, fx, color="coral") 
ax.plot(x, gx, color="olive")
ax.set_xlabel("osa x")
ax.set_ylabel("osa y")
ax.set_title("Funkce sin a cos")
```

### Nastavení os

```python
x = np.linspace(0, 8*np.pi, 100)
fx = np.sin(x)

fig, ax = plt.subplots()  
ax.plot(x, fx)
ax.axis([0, 40, 0, 1])
```

### Text

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots()

ax.hist(x, 20) # 20 bins
ax.text(60, 1200, r"$\mu = 115, \sigma = 15$") # klidne LaTeX v $...$,  r znaci raw string
```
